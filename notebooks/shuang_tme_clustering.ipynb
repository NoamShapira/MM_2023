```r
cnt <- read.csv('/home/labs/amit/noamsh/repos/MM_2023/outputs/composition_table_2024-03-26.csv')
dim(cnt)
head(cnt)
sort(cnt$Hospital.Code)
```

```r
options(repr.plot.width = 8, repr.plot.height = 8, repr.plot.res = 110)

hist(log10(cnt$TOTAL+1), breaks = 20)
cnt <- cnt[cnt$TOTAL>100,]
dim(cnt)
```

```r
col_pop_Imm <- c('chocolate4', 'lightsalmon', "purple", 'darkolivegreen', 'aquamarine3', 
                 'deepskyblue4', 'firebrick1', 'lightgreen', 'mediumseagreen', "magenta",
                 'darkturquoise', 'gold', 'darkred', 'deeppink', 'pink',
                 "skyblue", 'blue', 'darkblue', 'orange', 'grey50')
length(col_pop_Imm)
order_pop_Imm <- c("Mo", "Mo_CD16", "Neu_Pro", "DC_IRF8", "DC", 
                   "Mf", "Erythrocytes", "Mo_Pro", "B_Pro", "Mega", 
                   "Mast", "Fibro", "pDC", "B", "NK", 
                   "T_Naive", "T_Effector", "T_Effector_GZMB", "Doublets", "UN")
length(order_pop_Imm)

names(col_pop_Imm) <- order_pop_Imm
col_pop_Imm
```

```r
rownames(cnt) <- cnt$Hospital.Code.Time.Method
pct <- cnt[,setdiff(colnames(cnt), c('Hospital.Code.Time.Method', 'UN', 'TOTAL'))]
pct <- as.data.frame(t(apply(pct, 1, function(x) x/sum(x))))

col_pop_Imm <- col_pop_Imm[names(col_pop_Imm) %in% colnames(pct)]
pct <- pct[,names(col_pop_Imm)]
head(pct)
```

```r
options(repr.plot.width = 10, repr.plot.height = 10, repr.plot.res = 110)

library(ComplexHeatmap)
HM <- Heatmap(cor(t(pct)))
HM
```

```r
row_dendrogram <- row_order(HM)
rownames(pct)[row_dendrogram]
```

```r
library(reshape2)
library(ggpubr)
options(repr.plot.width = 32, repr.plot.height = 9, repr.plot.res = 110)

dat <- pct
dat$Hospital.Code.Time.Method <- cnt$Hospital.Code.Time.Method
res_pct <- melt(dat)
head(res_pct)
PID_ann <- rownames(pct)[row_dendrogram]
res_pct$Hospital.Code.Time.Method <- factor(res_pct$Hospital.Code.Time.Method, levels = PID_ann)

ggbarplot(res_pct, "Hospital.Code.Time.Method", "value",
  fill = "variable", color = "variable", palette = col_pop_Imm, xlab = 'Patient.ID', ylab = 'Fraction') + rotate_x_text()
```

```r
ha = HeatmapAnnotation(Type = anno_barplot(pct[PID_ann,],
                                          bar_width = 1, height = unit(12, "cm"),
                                          gp = gpar(fill = col_pop_Imm)),
                       Method = PID_meta$Method,
                       Time = PID_meta$Time,
                       col = list(Method = c("MARS" = "steelblue", "SPID" = "lightgreen"),
                                  Time = c("Pre" = "black", "Post" = "brown", "C1D22" = "red", "C3D1" = "orange", "nan" = "grey"))
                      )
# draw(ha)
```

```r
mat <- matrix(0, ncol = length(PID_ann), nrow = 1)
colnames(mat) <- PID_meta$Hospital.Code
Heatmap(mat, col = 'white',
        cluster_columns = F,
        bottom_annotation = ha)
```
